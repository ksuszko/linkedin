```python
import pandas as pd
import numpy as np
import csv
```

```python
References=pd.read_csv('References.csv')
df_r=pd.DataFrame(References)

print(df_r['REFERENCE_NAME'])
```

```python
#remove nulls
df_r_1=df_r[~df_r['REFERENCE_NAME'].isnull()]
#print(df_r_1)
df_r_1=df_r_1.astype(str)

#equal dataframe
df_r_2=df_r_1

#remove mr. and ms.
df_r_1['REFERENCE_NAME'] = df_r_1['REFERENCE_NAME'].str.replace('Mr. ','')
#remove mr. and ms.
df_r_1['REFERENCE_NAME'] = df_r_1['REFERENCE_NAME'].str.replace('Ms. ','')

#get first names
first_names = df_r_1.apply(lambda x: x.str.split().str[0])
df_r_2['First Name']=first_names['REFERENCE_NAME']
#print(first_names)
#print(df_r_2)

#get last names
last_names = df_r_1.apply(lambda x: x.str.split().str[1])
df_r_2['Last Name']=last_names['REFERENCE_NAME']
#print(last_names['REFERENCE_NAME'])
#print(df_r_2)

#get title (1 comma)
title = df_r_1.apply(lambda x: x.str.split(',').str[1])
df_r_2['Title']=title['REFERENCE_NAME']
#print(df_r_2['Title'])

#get department (2 comma)
department = df_r_1.apply(lambda x: x.str.split(',').str[2])
df_r_2['Dept']=department['REFERENCE_NAME']
#print(df_r_2['Dept'])
```

```python
#remove commas from names
df_r_2['First Name'].replace(',','', regex=True, inplace=True)
df_r_2['Last Name'].replace(',','', regex=True, inplace=True)

#remove DO, NOT, Who? Nan, Reference 
df_r_2['First Name'] = df_r_1['First Name'].str.replace('DO','')
df_r_2['First Name'] = df_r_1['First Name'].str.replace('NOT','')
df_r_2['First Name'] = df_r_1['First Name'].str.replace('Who?','')
df_r_2['First Name'] = df_r_1['First Name'].str.replace('Reference','')
df_r_2['First Name'] = df_r_1['First Name'].str.replace('REFERENCE','')
df_r_2['First Name'] = df_r_1['First Name'].str.replace('NO','')
df_r_2['First Name'] = df_r_1['First Name'].str.replace('reference','')
df_r_2['First Name'] = df_r_1['First Name'].str.replace('Plus','')
df_r_2['First Name'] = df_r_1['First Name'].str.replace('?','')

df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('DO','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('NOT','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('not','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('gone','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('retired','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('Retired','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('Who?','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('Reference','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('REFERENCE','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('reference','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('Plus','')
df_r_2['Last Name'] = df_r_1['Last Name'].str.replace('?','')

print(df_r_2)
df_r_2.to_csv('References_clean.csv')
```
